### 3)

The primal problem is $$\left(\mathbf{w}^{*}, w_{0}^{*}, \xi^{*} \in \mathbb{R}^{n}\right) \in \underset{\mathbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}, \xi \in \mathbb{R}^{n}}{\arg \min }\left(\frac{1}{2}\|\mathbf{w}\|^{2}+C \sum_{i=1}^{n} \xi_{i}\right)$$ with constraits $$\begin{array}{ll}{\xi_{i} \geq 0,} & {\forall i \in\{1, \cdots, n\}} \\ {y_{i}\left(\left\langle\mathbf{w}, \Phi\left(\mathbf{x}_{i}\right\rangle+ w_{0}\right) \geq 1-\xi_{i},\right.} & {\forall i \in\{1, \cdots, n\}}\end{array}$$ So we can have $$\xi_{i}=\max \left\{0,1-y_{i}\left(<w, \phi\left(x_{i}\right)>+w_{0}\right)\right\}$$ Then we put this $\xi_{i}$ into the primal formula, we can get: $$\underset{\mathbf{w} \in \mathcal{H}, w_{0} \in \mathbb{R}}{\arg \min }\left(\frac{1}{2}\|\mathbf{w}\|^{2}+C \sum_{i=1}^{n}\left[1-y_{i}\left(\left\langle\mathbf{w}, \Phi\left(\mathbf{x}_{i}\right)\right\rangle+ w_{0}\right)\right]_{+}\right)$$

### 4)

In this separation problem of SVM, the classification error is measured by misclassified points, and the margin is positive when a point is correctly classified and negative when it is not. By just set a value 1 to misclassified points, we can count the numbers thus calculate the error, but this count is not continous and not convex, so we cannot do the optimization, by applying the *Hinge* function, we have a convex version which will keep a similiar value, thus the optimization of this function minimizes the classification error.

It is observed that as the parameter C is decreasing, though the predication accuraray calculated holds the same high value, the lesser class will no longer be well separated.

## Classification of faces

Load the faces data:

```python
from time import time
import pylab as pl

from sklearn.cross_validation import train_test_split
from sklearn.datasets import fetch_lfw_people

# Download the data (if not already on disk); load it as numpy arrays
lfw_people = fetch_lfw_people(min_faces_per_person=70, resize=0.4,
                              color=True, funneled=False, slice_=None,
                              download_if_missing=True)
# data_home='.'

# introspect the images arrays to find the shapes (for plotting)
images = lfw_people.images / 255.
n_samples, h, w, n_colors = images.shape

# the label to predict is the id of the person
target_names = lfw_people.target_names.tolist()
```

```python
print(target_names)
```

```python
print("Number of samples: " + str(n_samples))
```

Pick a pair of data to classify:

```python
names = ['George W Bush', 'Gerhard Schroeder']
# names = ['Donald Rumsfeld', 'Colin Powell']

idx0 = (lfw_people.target == target_names.index(names[0]))
idx1 = (lfw_people.target == target_names.index(names[1]))
images = np.r_[images[idx0], images[idx1]]
n_samples = images.shape[0]
y = np.r_[np.zeros(np.sum(idx0)), np.ones(np.sum(idx1))].astype(np.int)
```

Split the data:

```python
# features using only illuminations
X = (np.mean(images, axis=3)).reshape(n_samples, -1)

# # or compute features using colors (3 times more features)
# X = images.copy().reshape(n_samples, -1)

# Scale features
X -= np.mean(X, axis=0)
X /= np.std(X, axis=0)

indices = np.random.permutation(X.shape[0])
train_idx, test_idx = indices[:int(X.shape[0] / 2)], indices[int(X.shape[0] / 2):]
X_train, X_test = X[train_idx, :], X[test_idx, :]
y_train, y_test = y[train_idx], y[test_idx]
images_train, images_test = images[
    train_idx, :, :, :], images[test_idx, :, :, :]
```

Quantitative evaluation of the model quality on the test set:

```python
t0 = time()
# Fit a classifier, 
clf = SVC(kernel='linear')
clf.fit(X_train,y_train)

t0 = time()

# Predict labels for the X_test images
y_pred = clf.predict(X_test)

print("done in %0.3fs" % (time() - t0))
print("Chance level : %s" % max(np.mean(y), 1. - np.mean(y)))
print("Accuracy : %s" % clf.score(X_test, y_test))
```

### 5)

```python
error = []

num = 11
C= np.logspace(-5,5,num)
for i in range(0,11):
    clf1 = SVC(kernel='linear', C = C[i])
    clf1.fit(X_train, y_train)
    error.append(1 - clf1.score(X_test, y_test))
    
plt.figure()
plt.plot(C, error)
plt.xlabel("regularization parameter C ")
plt.ylabel("error")
plt.xscale("log")
plt.show()
```

We can see that as parameter C gets larger, the error gets smaller.

```python
def plot_gallery(images, titles, n_row=3, n_col=4):
    """Helper function to plot a gallery of portraits"""
    pl.figure(figsize=(1.8 * n_col, 2.4 * n_row))
    pl.subplots_adjust(bottom=0, left=.01, right=.99, top=.90,
                       hspace=.35)
    for i in range(n_row * n_col):
        pl.subplot(n_row, n_col, i + 1)
        pl.imshow(images[i])
        pl.title(titles[i], size=12)
        pl.xticks(())
        pl.yticks(())


def title(y_pred, y_test, names):
    pred_name = names[int(y_pred)].rsplit(' ', 1)[-1]
    true_name = names[int(y_test)].rsplit(' ', 1)[-1]
    return 'predicted: %s\ntrue:      %s' % (pred_name, true_name)

prediction_titles = [title(y_pred[i], y_test[i], names)
                     for i in range(y_pred.shape[0])]

plot_gallery(images_test, prediction_titles)
pl.show()
```

Because Support Vector Machine algorithms are not scale invariant, if we don't center and reduce the features, the algorithm will not be effective across a range of inputs and applications.

### 8)

```python
from sklearn import decomposition

clf_rbf = SVC(kernel='rbf', C=1.0)
clf_rbf.fit(X_train, y_train)

predict = clf_rbf.predict(X_test)
print("The score before PCA:")
print(clf_rbf.score(X_test, y_test))


n_components = 150
pca = decomposition.RandomizedPCA(n_components=n_components, whiten=True).fit(X_train)
X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)

clf_rbf = SVC(kernel='rbf',class_weight='balanced')
clf_rbf.fit(X_train_pca, y_train)
predict = clf_rbf.predict(X_test_pca)

print("Score after PCA:")
print(clf_rbf.score(X_test_pca, y_test))
```

We can see that the performance of the RBF kernel (about 0.78) is not as good as the linear one (about 0.94), evne after dimension reduction (about 0.89). The RBF kernel will cause overfitting, but after the reduction of dimension, this problem is solved.

## To go further: Calculation of duality gap

```python
num = 11
tol = np.logspace(-5, 0, num)
Y_train = 2*y_train-1

vP_List = []
vD_List = []
diff_List = []

for n in range(num):
    clf = svm.SVC(kernel='linear', tol = tol[n])
    clf.fit(X_train, Y_train)
    w = clf.coef_[0]
    
    # pour le problème primal (C=1)
    vP = 0.5* np.dot(w.T, w)
    for i in range(X_train.shape[0]):
        middle = 1 - Y_train[i] * (np.dot(w, X_train[i]) + clf.intercept_[0])
        vP = vP + np.amax([middle, 0])
    vP_List.append(vP)
   

    # lpour le problème dual
    dualCoef = clf.dual_coef_[0]
    alpha = dualCoef * np.r_[(-1) * np.ones(clf.n_support_[0]), np.ones(clf.n_support_[1])]
    vD = np.sum(alpha)
    for i in range(len(dualCoef)):
        for j in range(len(dualCoef)):
            vD = vD - 0.5 * dualCoef[i] * dualCoef[j] * np.dot(clf.support_vectors_[i], clf.support_vectors_[j])
    vD_List.append(vD)  
    diff_List.append(vP - vD)

plt.figure()
plt.plot(tol, vP_List, 'o-', color="y", label = "primal value")
plt.plot(tol, vD_List, 'x-', color="r", label = "duale value")
plt.xscale("log")
plt.xlabel("Tolerance")
plt.ylabel("Value")
plt.legend(loc="best")
```

As the tolerance increases, the difference between the two values become bigger.
