```python
# Importing modules
import pandas as pd
from sklearn.metrics import accuracy_score
import numpy as np
from sklearn.model_selection import train_test_split
```

Data Pre-pocessing

```python
data = pd.read_csv('data.csv', delimiter=';', error_bad_lines=False)
data
```

```python
df = pd.read_csv('data.csv', delimiter=';', error_bad_lines=False)
df['quality'] = pd.cut(data['quality'], [0,4,5,7,10], labels = ['Low','Medium-low','Medium-high','High'])

print(df)
```

```python
data_train, data_test = train_test_split(df, test_size=0.2)
print(data_train)
print(data_test)

```

```python
Y_train = data_train.quality
Y_train.reset_index(drop=True, inplace=True)
Y_test = data_test.quality
Y_test.reset_index(drop=True, inplace=True)
print(Y_train)
print(Y_test)

```

```python
X_train = data_train.drop(columns=['quality'])
X_train.reset_index(drop=True, inplace=True)
X_test = data_test.drop(columns=['quality'])
X_test.reset_index(drop=True, inplace=True)
print(X_train)
print(X_test)

```

```python
X_train.shape
X_train = X_train.values.reshape(-1,11)
X_train
```

```python
X_test.shape
X_test = X_test.values.reshape(-1,11)
X_test
```

```python
from sklearn.naive_bayes import MultinomialNB
mnb=MultinomialNB()
mnb.fit(X_train,Y_train)
y_predict=mnb.predict(X_test)

```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline

model = make_pipeline(TfidfVectorizer(), MultinomialNB())

model.fit(X_train, Y_train)
labels = model.predict(X_test)
```

```python
from sklearn import metrics
metrics.precision_score(X_train, Y_train, average='micro') 
```
